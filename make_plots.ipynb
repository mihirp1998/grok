```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm

print(fm.fontManager.ttflist)
```

```python
# set font size and style
plt.rcParams.update({'font.size': 12})
plt.rcParams["font.family"] = "Times New Roman"
# plt.rcParams["font.family"] = "Palatino"
```

```python
ff  = [10,100,100,100,100,100,100,100,100]
data_samples = [10,20,30,40,50,60,70,80,90]

ff_acc = np.array(ff)
data_samples = np.array(data_samples)
```

```python
plt.figure(dpi=250, figsize=(4, 3))

plt.plot(data_samples, ff_acc, label='Ours', c='C0', marker='o')
# plt.plot(num_params, brn_loss, label='Block Reversible', c='C1', marker='s')
# plt.xlim(0, 100)
# plt.ylim(0, 110)
plt.legend()
plt.title('Operation: +')
plt.xlabel('Train Data Fraction (%)')
plt.ylabel('Validation Accuracy')
plt.grid()
plt.tight_layout(pad=0)
# plt.savefig('id_vs_ood_accs.pdf', bbox_inches='tight')
plt.show()
```

```python
ff  = [3.05,1.53,2.2,1.83,0.48]
brn = [3.0,2.05,0.2,0.1,0.03]
num_layers = [6,12,24,48,96]

ff_loss = np.array(ff)
brn_loss = np.array(brn)
num_layers = np.array(num_layers)
```

```python
plt.figure(dpi=250, figsize=(4, 3))

plt.plot(num_layers, ff_loss, label='Forward Forward', c='C2', marker='o')
plt.plot(num_layers, brn_loss, label='Block Reversible', c='C1', marker='s')

plt.legend()
plt.xlabel('Number of Blocks (Depth)')
plt.ylabel('Negative Log LikeLihood Loss')
plt.grid()
plt.tight_layout(pad=0)
# plt.savefig('id_vs_ood_accs.pdf', bbox_inches='tight')
plt.show()
```
